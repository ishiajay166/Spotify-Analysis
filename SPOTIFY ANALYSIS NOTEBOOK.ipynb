```python
#import libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
#reading csv tracks
df_tracks = pd.read_csv('C:\\Users\\Ishika\\Desktop\\Desktop main\\spotify project\\tracks.csv')
df_tracks.head()
```

```python
# checking for sum of null values in each column
pd.isnull(df_tracks).sum()
```

```python
#display rows with NaN values
df_tracks_NaN = df_tracks[df_tracks.isnull().any(axis=1)]
df_tracks_NaN.head()
```

```python
# we could either look for the missing data but here just droping the rows with nan values 
for i in df_tracks_NaN['id'].tolist():
    df_tracks=df_tracks[df_tracks['id']!=i]
print(df_tracks.isnull().sum())
```

```python
#checking information of tracks
df_tracks.info()
```

```python
# to set column release_date as index and changing its datatype
df_tracks.set_index('release_date',inplace=True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
#check the artist at the 18th index
df_tracks[["artists"]].iloc[18]
```

```python
# coverting the column with milisecinds in just seconds 
df_tracks['duration']=df_tracks['duration_ms'].apply(lambda x: round(x/1000))
df_tracks.drop('duration_ms',inplace=True,axis=1)
df_tracks.head()
```

```python
# correlation plot of tracks after excluding columns key,mode,explicit
corr_df=df_tracks.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g", vmin=-1,vmax=1,center=0,cmap="inferno",linewidths=1,linecolor="Black")
heatmap.set_title("correlation Heatmap between variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
#take just 0.4 percent of data 
sample_df=df_tracks.sample(int(0.004*len(df_tracks)))
sample_df
```

```python
#length of sample
print(len(sample_df))
```

```python
# regressing plot of loudness and energy
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="loudness",x="energy",color="c").set(title="loudness vc energy corr")
```

```python
# regression plot of popularity and acousticness
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="popularity",x="acousticness",color="r").set(title="popularity vs acousticness corr")
```

```python
#creating column year date from index and extracting year
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
year=df_tracks.dates.dt.year
year
```

```python
#update seaborn
#pip install --user seaborn==0.11.0
```

```python
#ditribution plot of data with respect to year
sns.displot(year,discrete=True,aspect=2,height=5,kind="hist").set(title="number of sings per year")# aspect gives the width=aspect multiplied with height
```

```python
# see the duration of songs over the years 
total_dr=df_tracks.duration
fig, ax=plt.subplots(figsize=(18,7))
fig=sns.barplot(x=year,y=total_dr,ax=ax,errwidth=False).set(title="yer vs duation")
plt.xticks(rotation=90)
```

```python
#line plot of duration with respect to year
total_dr=df_tracks.duration
sns.set_style(style="whitegrid")
fig,ax=plt.subplots(figsize=(10,5))
fig=sns.lineplot(x=year,y=total_dr,ax=ax).set(title="year vs duration")
plt.xticks(rotation=60)
```
